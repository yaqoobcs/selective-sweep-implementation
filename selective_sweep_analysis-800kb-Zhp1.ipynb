```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path = r'/home/ibge/fnc_files/'
# Read the data from a file into a pandas DataFrame
data = pd.read_csv(path+'BL_HS_MZ_GJ_TR_WZ_MC_fnc.csv', delimiter='\t')
```

```python

# Calculate mean values
data['mean'] = (data['BL_AF'] + data['HS_AF'] + data['MZ_AF'] + data['GJ_AF'] + data['TR_AF'] + data['WZ_AF'] + data['MC_AF']) / 7
```

```python
output_name = 'win_800kb_Zhp_fattial-0.6'

# Define window parameters
window_size = 800000  # in base pairs (150kb)
overlap = 0.5  # 50% overlap

win_hp = []
chromosomes = []
win_no = []
w_no=0
win_onset = []
win_end = []
win_s = []
win_l = []

# Slide the window over the data
start_position = 0
end_position = start_position + window_size
```

```python
chrom_27 = data['CHROM'].unique()
```

```python
chrom_27
```

```python
for ch in chrom_27:
    ch_df = data[(data['CHROM'] == ch)] 
    ch_min = ch_df['POS'].min()
    ch_max = ch_df['POS'].max()
    
    start_position = ch_min
    end_position = start_position + window_size
    
    while start_position < ch_max:

        window_data = data[(data['POS'] >= start_position) & (data['POS'] < end_position) & (data['CHROM'] == ch)]

    
        if len(window_data) > 10:
            w_no+=1
            chrom_n = window_data['CHROM'].iloc[0]
            start_pos_chr = window_data['POS'].iloc[0]
            end_pos_chr = window_data['POS'].iloc[-1] 
            
            
            w_hp = 2*((window_data['mean'] > 0.6).sum() * (window_data['mean'] < 0.4).sum())/ ((window_data['mean'] > 0.6).sum() + (window_data['mean'] < 0.4).sum())**2
            
            
            win_hp.append(w_hp)
            win_no.append(w_no)
            chromosomes.append(chrom_n)
            win_onset.append(start_pos_chr )
            win_end.append(end_pos_chr )
            win_s.append(end_pos_chr - start_pos_chr)
            win_l.append(len(window_data))
    
        start_position = int(start_position + window_size * (1 - overlap))
        end_position = int(start_position + window_size)
    print(ch)

# Create a DataFrame to store the results
result_df = pd.DataFrame({'win_no': win_no, 'chrom_n': chromosomes, 'win_onset': win_onset, 'win_end':win_end, 'win_size':win_s, 'no_SNP':win_l, 'win_hp':win_hp})
```

```python
# Calculate the mean and standard deviation of ZHp values
mean_Hp = result_df['win_hp'].mean()
stddev_Hp = result_df['win_hp'].std()
```

```python
mean_Hp, stddev_Hp
```

```python
result_df['Zhp'] = (result_df['win_hp'] - mean_Hp) / stddev_Hp
```

```python
result_df.to_csv(output_name +'.csv', index=False)
```

```python
result_df
```

```python
# Create a dictionary to map unique numeric values to each unique chromosome name
chrom_to_numeric = {chrom: idx for idx, chrom in enumerate(result_df['chrom_n'].unique())}
```

```python

# Group data by chromosome
chromosome_groups = result_df.groupby('chrom_n')

# Create a color map for alternating colors
colors = ['k', 'r']

# Create a color map with 26 different colors
#colors = plt.cm.get_cmap('tab20', 20).colors


# Plotting
plt.figure(figsize=(15, 5))
for chrom, value in chrom_to_numeric.items():
    color = colors[value % 2]  # Select color based on chromosome number
    #color = colors[value % 20]
    c_df = result_df[result_df['chrom_n'] == chrom]
    plt.scatter(c_df['win_no'], c_df['Zhp'], marker='.', label=f'{chrom}', color=color)



plt.xlabel('win_no')
plt.ylabel('Z Transformation')
plt.title('Z Transformation of Pooled Heterozygosity')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.axhline(-5, color='b', linestyle ='--', alpha = 0.5)
plt.savefig(output_name+'_.pdf', bbox_inches='tight')
plt.show()
```
